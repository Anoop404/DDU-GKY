# Grouping and Aggregation:

### groupby()

```python
Sex_Survived_sum=df.groupby(['Sex'])['Survived'].sum()
```

```python
Sex_Survived_sum
```

```python
Sex_Survivied_Avg = df.groupby(['Sex'])['Survived'].mean()*100
```

```python
Sex_Survivied_Avg
```

```python
#Q1: Total number of Males and Females boardrd from each port
```

```python
Sex_Survived_Board=df.groupby(['Sex','Embarked'])['Survived'].sum()
Sex_Survived_Board
```

```python
#Q2: Total amount of Fare collected from Males and Females boarded from each port.
```

```python
Total_Fare_Collected = df.groupby(['Sex','Embarked'])['Fare'].sum()
Total_Fare_Collected
```

```python
#Q3: Total amount fare collected from each port.
```

```python
Fare_Collected = df.groupby(['Embarked'])['Fare'].sum()
Fare_Collected
```
